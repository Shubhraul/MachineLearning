```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import MinMaxScaler, Normalizer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
import warnings
warnings.filterwarnings(action='ignore')
```

```python
processed_data = pd.DataFrame(data.loc[:,'Present_Price'])
# Ordinal Encoding on Year
Ordinal_Encoder = OrdinalEncoder(categories=[list(sorted(data['Year'].unique()))])
processed_data['Year'] = Ordinal_Encoder.fit_transform(data[['Year']])
#Scaling the Kms Countinous 
Scaler = MinMaxScaler()
processed_data['Kms_Driven'] = Scaler.fit_transform([data['Kms_Driven']])[0]
# One Hot Encoding on Categorical Data
OneHot = OneHotEncoder()
processed_data[OneHot.get_feature_names_out()] = pd.DataFrame(OneHot.fit_transform(data[['Fuel_Type','Seller_Type','Transmission','Owner']]).toarray())
processed_data
```

```python
actual_price=processed_data.iloc[:,0]
features=processed_data.iloc[:, 1:]
features
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, actual_price, test_size=0.2, random_state=1)
df = pd.DataFrame() 
```

```python
def report(Y_test,Y_pred,model):
    r2_score_model= round(r2_score(Y_test,Y_pred),2)
    root_mean_square= round(math.sqrt(mean_squared_error(Y_test,Y_pred)),2)
    mean_abs_error=round(mean_absolute_error(Y_test,Y_pred),2)
    mean_absolute_error_percentage= round(mean_absolute_percentage_error(Y_test,Y_pred),2)
    print(model)
    print("R2_Score",r2_score_model)
    print("Root Mean Square",root_mean_square)
    print("Mean Absolute error",mean_abs_error)
    print("Mean Absolute error percentage",mean_absolute_error_percentage)
    dict1 = {
        'Model' : model,
        'R2 Score' : r2_score_model,
        'Root Mean Square Error' : root_mean_square,
        'Mean Absolute Error' : mean_abs_error,
        'Mean Absolute Percentage Error' : mean_absolute_error_percentage
    }
    return dict1 
```

```python
Linear_Model = LinearRegression()
Linear_Model.fit(X_train,y_train)
y_pred = Linear_Model.predict(X_test)
df1 = pd.DataFrame([report(y_test,y_pred,'LinearRegression')])
df = pd.concat([df,df1])
```

```python
Ridge_Model = Ridge()
Ridge_Model.fit(X_train,y_train)
y_pred = Ridge_Model.predict(X_test)
df1 = pd.DataFrame([report(y_test,y_pred,'RidgeRegression')])
df = pd.concat([df,df1])
```

```python
RandomForest_Model = RandomForestRegressor()
RandomForest_Model.fit(X_train,y_train)
y_pred = RandomForest_Model.predict(X_test)
df1 = pd.DataFrame([report(y_test,y_pred,'Random Forest')])
df = pd.concat([df,df1])
```

```python
df.set_index(['Model'],inplace=True)
df
```

```python
fig = plt.figure(figsize=(12,6))
sns.lineplot(df)
plt.tight_layout()
```

```python
Model = {
    "Linear" : LinearRegression(),
    "Ridge" : Ridge(),
    "RandomForest" : RandomForestRegressor(),
    'KNN Regression' : KNeighborsRegressor(n_neighbors=4)
}

Pearameter = {
    "Ridge" : {
        'solver' : ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga', 'lbfgs'],
        'alpha' : [0.1,0.01,0.001,1]
    },
    "RandomForest" : {
         'criterion' : ["squared_error", "friedman_mse", "absolute_error", "poisson"],
        'bootstrap': [True, False],
        'max_features': ["sqrt", "log2", None],
    },
    'KNN Regression' : {
        'n_neighbors':[i for i in range(2,15)],
        'weights':['uniform','distance'],
        'p':[1,2],
        'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute']
    }
}
```

```python
Best_Score,Best_Parameter=0,{'Linear Regression' : None}
for i , mod in Model.items():
    if(i=='Linear'): continue
    grid = GridSearchCV(estimator=mod,param_grid=Pearameter[i],cv=5);
    grid.fit(X_train,y_train)
    Best_Parameter[i] = grid.best_params_
    print('For '+i+' Model ')
    print('Best Score is ',grid.best_score_)
    print('Parameter at best score is ',grid.best_params_,end='\n\n')
```

```python
Best_Parameter
```

```python
Ridge_Model = Ridge(**Best_Parameter['Ridge'])
Ridge_Model.fit(X_train,y_train)
y_pred = Ridge_Model.predict(X_test)
df1 = pd.DataFrame([report(y_test,y_pred,'RidgeRegression')])
```
