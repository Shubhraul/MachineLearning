```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score,accuracy_score
from sklearn.model_selection import StratifiedKFold
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.neighbors import KNeighborsClassifier
from yellowbrick.classifier import ROCAUC
from yellowbrick.classifier.rocauc import roc_auc
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
car_data = pd.read_csv('car_evaluation.csv')
car_data.columns = ['Buying_price','Maintainance_Price','No_of_doors','Person_Capacity','Size_of_luggage','Safety','Acceptability']
car_data.head()
```

```python
car_data.isna().sum()
```

```python
car_data.dtypes
```

### Data Visualization 

```python
fig = plt.figure(figsize=(12,14))
fig.suptitle('Stacked plot of Car Features with Acceptability')
for i in range(1,7):
    column = car_data.columns[i-1]
    ax = fig.add_subplot(3,2,i)
    group = car_data[['Acceptability',column]].groupby(['Acceptability',column]).size().unstack(fill_value=0)
    group.plot(kind='bar',stacked=True, ax=ax)
    ax.set_title('Frequency of '+column+' with Target Acceptability')
    plt.tight_layout()
```

### Encoding 

```python
##Label Encoding On Features
Encoded_car = pd.DataFrame()
Encoder = LabelEncoder()
for i in car_data.columns[:-1]:
    Encoded_car[i] = Encoder.fit_transform(car_data[i])
    
##Ordinal Encoding on Target Class
categry = car_data['Acceptability'].unique()
encoder = OrdinalEncoder(categories=[categry])
Target = pd.DataFrame(encoder.fit_transform(car_data[['Acceptability']]),columns=['Acceptability'])
```

```python
cor = pd.concat([Encoded_car,Target],axis=1).corr()['Acceptability'].drop('Acceptability').sort_values()
sns.barplot(x=cor.values,y=cor.index,orient='h').set_title('Correlation of Features with Acceptability')
```

```python
x_train, x_test, y_train, y_test = train_test_split(Encoded_car,Target,test_size=0.2 , random_state=100)
```

### Model Implementation 

```python
Class = pd.DataFrame()
def report(model,Y_test,Y_pred):
    cls = {
        'Accuracy' : round(accuracy_score(Y_test,Y_pred),2),
    'Precision Wgt' : round(precision_score(Y_test,Y_pred,average='weighted'),2),
    'F1 Score Wgt': round(f1_score(Y_test,Y_pred,average='weighted'),2),
     'Recall Wgt': round(recall_score(Y_test,Y_pred,average='weighted'),2)
    }
    return pd.DataFrame(cls.values(),index=cls.keys(),columns=[model])

def roc(model):
    visual = ROCAUC(model)
    visual.fit(x_train,y_train)
    visual.score(x_test,y_test)
    visual.show()
    return

def Model(model_key,cls,paras=None,ratio=0):
    if(paras): cls.set_params(**paras)
    cls.fit(x_train,y_train)
    y_pred = cls.predict(x_test)
    test_str = '_'+str(ratio) if ratio > 0 else ''
    rep = report(model_key+test_str,y_test,y_pred)
    print(rep)
    global Class
    Class = pd.concat([Class,rep],axis=1)
    sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d',cmap='rocket_r',xticklabels=categry,yticklabels=categry).set_title(model_key+' Confussion Matrix')
    return cls
```

```python
Classifiers = {
    'Logistic Regression' : LogisticRegression(),
    'Naive Bayes' : GaussianNB(),
    'Decision Tree' : DecisionTreeClassifier(),
    'KNN Classifier' : KNeighborsClassifier(), 
    'Random Foreast' : RandomForestClassifier(),
    'Gradien Boost' : GradientBoostingClassifier(),
}

Parameters = {
    'Logistic Regression': {
        'penalty' : ['l1', 'l2', 'elasticnet', None],
        'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
        'C' : list(np.arange(0.1,2,0.1))
    },
    'Naive Bayes' : {
        'var_smoothing': np.logspace(0,-9, num=100),
    },
    'Decision Tree' : {
        'criterion' : ["gini", "entropy", "log_loss"],
        'splitter' : ["best", "random"],
    },
    'KNN Classifier' : {
        'n_neighbors':[i for i in range(2,15)],
        'weights':['uniform','distance'],
        'p':[1,2]
    },
    'Random Foreast' : {
        'criterion' : ["gini", "entropy", "log_loss"],
        'bootstrap': [True, False],
        'max_features': ['auto', 'sqrt'],
    },
    'Gradien Boost' : {
        'loss' : ['log_loss', 'deviance', 'exponential'],
        'criterion' : ['friedman_mse', 'squared_error'],
        'max_features' : ['auto', 'sqrt', 'log2'],
    }
}
```

```python
fig,j,Trained_Model = plt.figure(figsize=(12,6)),1,[]
for key, mod in Classifiers.items():
    ax = fig.add_subplot(2,3,j)
    Trained_Model.append(Model(key,mod))
    j+=1
    plt.tight_layout()
```

### ROC Curve

```python
for i in Trained_Model: roc(i)
```

### Decision Tree plot

```python
plot_tree(Trained_Model[2], feature_names=Encoded_car.columns, class_names=categry, filled=True)
plt.show()
```

### Hyper Parameter Tuning 

```python
Best_Score,Best_Parameter=0,{}
for i , mod in Classifiers.items():
    grid = GridSearchCV(estimator=mod,param_grid=Parameters[i],cv=5);
    grid.fit(x_train,y_train)
    Best_Parameter[i] = grid.best_params_
    print('For '+i+' Model ')
    print('Best Score is ',grid.best_score_)
    print('Parameter at best score is ',grid.best_params_,end='\n\n')
```

### Tuned Model implementaion for different Test Ratio

```python
Trained_Model = []
for i in [0.1,0.2,0.3]:
    x_train, x_test, y_train, y_test = train_test_split(Encoded_car,Target,test_size=i, random_state=100)
    fig,j = plt.figure(figsize=(12,6)),1
    fig.suptitle('For Test Ratio of '+str(i))
    for key, mod in Classifiers.items():
        ax = fig.add_subplot(2,3,j)
        Trained_Model.append(Model(key,mod,Best_Parameter[key],i))
        j+=1
    print(end='\n\n\n')
```

### Model Evaluation 

```python
fig2 = plt.figure(figsize=(10,16))
for i in range(0,4):
    ax2 = fig2.add_subplot(4,1,i+1) 
    STR = 'With Parameter Tuned : TestSize '+str(i*10)+'%' if i !=0 else 'Without Tuning'
    sns.lineplot(Class.transpose().iloc[6*i:(6*i)+6,:]).set_title('Model Evaluation '+STR)
    plt.tight_layout()
```
