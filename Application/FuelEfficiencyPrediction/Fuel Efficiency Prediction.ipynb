## <center>Fuel Efficiency Prediction</center>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('auto-mpg.csv')
data.head()
```

```python
data.dtypes
```

```python
data.isna().sum()
```

```python
data.dropna(axis=0,inplace=True)
data.isna().sum()
```

```python
data.describe()
```

### Data Visualization

```python
sns.heatmap(data.corr(),annot=True).set_title('Correlation Matrix')
```

```python
fig = plt.figure(figsize=(10,12))
fig.suptitle('Feature Visualization with mpg')
for i in range(8):
    ax = fig.add_subplot(4,2,i+1)
    sns.scatterplot(data,x=data.columns[i],y='mpg').set_title(data.columns[i]+' VS mpg')
    plt.tight_layout()
```

```python
fig1 = plt.figure(figsize=(10,12))
fig1.suptitle('Distribution Plot')
for i in range(8):
    ax1 = fig1.add_subplot(4,2,1+i)
    sns.histplot(data.iloc[:,i],kde=False).set_title(data.columns[i])
    plt.tight_layout()
    plt.grid()
```

### Encoding 

```python
Encoded_data = data.loc[:,['displacement','horsepower','weight','acceleration']]
OneHot_Encoder = preprocessing.OneHotEncoder()
Encoded_data[OneHot_Encoder.get_feature_names_out()] = pd.DataFrame(OneHot_Encoder.fit_transform(data[['cylinders','model year','origin']]).toarray())
```

```python
Feature = data.drop(['mpg','car name'],axis=1)
Target = data[['mpg']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(Feature, Target, test_size=0.2, random_state=100)
```

### Linear Regression

```python
Model = LinearRegression()
Model.fit(x_train,y_train)
Model.score(x_test,y_test)
```
