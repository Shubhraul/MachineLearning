```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mentioned_data = pd.read_excel('Tunning_MentionedFeatures.xlsx')
mentioned_data.head()
```

```python
def annote(ax,h=0):
    for p in ax.patches:
        ax.text(p.get_x() + p.get_width() / 2,
                p.get_height()+h,
                f'{p.get_height()}',
                ha='center',
                va='bottom' 
               )
    return None
```

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1,1,1)
sns.barplot(data=mentioned_data, x='Model Name', y='Accuracy',hue='State').set_title('Mentioned Features')
ax = plt.gca()
annote(ax)
plt.savefig('Mentined feature.png')
```

```python
Tunned_DT = pd.read_excel('Tunning_DT_15.xlsx')
Tunned_DT.head()
```

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1,1,1)
sns.barplot(data=Tunned_DT, x='Model Name', y='Accuracy',hue='State').set_title('Feature Selection using Decision Tree')
ax = plt.gca()
annote(ax)
plt.savefig('Decision Tree feature.png')
```

```python
Tunned_RF = pd.read_excel('Tunning_RF_15.xlsx')
Tunned_RF.head()
```

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1,1,1)
sns.barplot(data=Tunned_RF, x='Model name', y='Accuracy',hue='State').set_title('Feature Selection using Random Forest')
ax = plt.gca()
annote(ax)
plt.savefig('Random Forest feature.png')
```
