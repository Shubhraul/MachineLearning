```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
# We are going to use Titanic Dataset for Encoding
Titanic = pd.read_csv('../Data/Titanic.csv')
Titanic.dtypes
```

```python
Titanic['Embarked'].value_counts()
```

```python
class_mapping = {
    1: '1st class',
    2: '2nd class',
    3: '3rd class'
}

Titanic['Passenger_Class'] = Titanic['Pclass'].map(class_mapping)

print(Titanic[['Pclass', 'Passenger_Class']].head())
class_mapping1 = {
    'C': 'Cherbourg',
    'Q': 'Queenstown',
    'S': 'Southampton'
}

Titanic['Embark'] = Titanic['Embarked'].map(class_mapping1)

print(Titanic[['Embarked', 'Embark']].head())
s = Titanic['Survived']
```

```python
Titanic.drop(['Pclass','Ticket','Cabin','Embarked','Survived'],axis=1,inplace=True)
```

```python
Titanic['Survived']=s
```

```python
Titanic.rename(columns={'Sex': 'Gender'}, inplace=True)
Titanic
```
